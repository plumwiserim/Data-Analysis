# 문제1-6) '지역별' 컬럼을 '시도'와 '구군'으로 구분해서 별도의 컬럼을 만드세요

# 문제1-7) '지역고용통계' 를 위한 전처리 함수를 작성하세요.

````python
def employment_preprocessing(path):
    employment_raw = pd.read_excel(employment_path, sheet_name = '데이터', skiprows=1)
    
    computational_work_index = employment_raw[employment_raw['산업별']!='전산업'].index
    computational_work = employment_raw.drop(computational_work_index)

    select_col = computational_work[['지역별', '전체종사자']]

    select_col.columns = ['지역명', '고용자수']

    select_col['시도'] = select_col['지역명'].dropna().apply(lambda x: x.split(' ')[0])
    select_col['구군'] = select_col['지역명'].dropna().apply(lambda x: x.split(' ')[1])
    return select_col
````

# 문제1-8) 전처리 함수를 사용해서 전처리된 결과를 출력하세요.

````python
employment_df = employment_preprocessing(employment_path)
employment_df
````

# 문제2-5) '주민등록세대수' 전처리 함수 만들기

````python
def people_preprocessing(path):
    people_raw = pd.read_excel(path, sheet_name = '데이터')
    people_raw.columns = ['시도', '구군', '세대']

    big_col = []
    for num, temp in enumerate(people_raw['시도']):
        if pd.isna(temp) :
            big_col.append(big_col[num-1])
        else:
            big_col.append(temp)

    def get_province(x):
        if len(x) != 4:
            return x[0:2]
        else:
            return x[0] + x[2]

    small_col = []
    for i in big_col:
        small_col.append(get_province(i))

    people_raw['시도'] = small_col

    sub_total_index = people_raw[people_raw['구군']=='소계'].index
    except_sub_total = people_raw.drop(sub_total_index)
    return except_sub_total
````

# 문제2-6) '주민등록세대수' 전처리 함수를 사용해서 출력

````python
people_df = people_preprocessing(people_path)
people_df
````
